```python
intersection = gpd.overlay(nui[["V5a", "geometry"]].to_crs(grid.crs), grid, how="intersection")
t2 = gpd.overlay(nui[["V5a", "geometry"]].to_crs(grid.crs), grid, how="difference")
```

```python

print(len(intersection))
```

```python
df_grid = grid
```

```python
df_grid["target"] = 0
intersection = intersection.rename({"V5a": "target"}, axis=1).drop(["geometry"], axis=1).drop_duplicates()
```

```python
intersection.columns
```

```python
df_grid.columns
```

```python
df_grid = df_grid.drop(["geometry"], axis=1)
```

```python
intersection = intersection.groupby('gid').apply(lambda group: group.sample(1)).reset_index(drop=True)
```

```python
missing_ids = pd.merge(intersection["gid"], df_grid, on='gid', how='right', indicator=True)
```

```python
len(missing_ids)
```

```python
result_df = missing_ids[missing_ids['_merge'] == 'right_only'].drop(['_merge'], axis=1)
multi_class_grid = pd.concat([intersection, result_df], ignore_index=True)
```

```python
len(multi_class_grid)
```

```python
multi_class_grid.to_csv("multi_class_grid.csv")
```
